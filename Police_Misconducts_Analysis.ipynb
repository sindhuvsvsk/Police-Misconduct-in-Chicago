### assign the date of 1990-01-01 to appointed_date with missing values

```python
r_list = []
for i in merged_set["appointed_date"]:
    if i == "unknown":
        i = "1990-01-01"
    r_list.append(i)
#print(len(r_list))
merged_set = pd.concat([merged_set, pd.DataFrame(r_list, columns = ["modified_appointed_date"], index = merged_set.index)], axis = 1)
```

### assign the date of 2018-12-31 to resignation_date with missing values

```python
r_list = []
for i in merged_set["resignation_date"]:
    if i == "unknown":
        i = "2018-12-31"
    r_list.append(i)
#print(len(r_list))
merged_set = pd.concat([merged_set, pd.DataFrame(r_list, columns = ["modified_resignation_date"], index = merged_set.index)], axis = 1)
#pd.DataFrame(r_list, columns = ["modified_resignation_date"], index = merged_set.index )
```

### Extract the year by the modified appointed date and resignation date

```python
merged_set['appointed_year'] = pd.DatetimeIndex(merged_set['modified_appointed_date']).year
merged_set['resignation_year'] = pd.DatetimeIndex(merged_set['modified_resignation_date']).year
merged_set
```

```python
merged_set = merged_set[merged_set.resignation_year >= 1990]
```

### Create a new column with the allegation count per year

```python
t = 1990
count_per_year = []
for i, j, c in zip(merged_set.appointed_year, merged_set.resignation_year, merged_set.allegation_count_after1990):
    if i < 1990:
        i = 1990
    dy = j - i
    if dy == 0:
        dy = 1
    count_per_year.append(round(c/dy, 2))
merged_set = pd.concat([merged_set, pd.DataFrame(count_per_year, columns = ["allegation_count_per_year"], index = merged_set.index)], axis = 1)
```

```python
top_offenders_per_year = merged_set.sort_values("allegation_count_per_year", ascending = False)
```

```python
top_offenders_per_year.allegation_count_per_year.head(20).plot(kind = "bar")
```
