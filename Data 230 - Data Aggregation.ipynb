```python
# Create a function to add new column to indicate whether that police officer served in each individual year
def new_served_year_column(year, name):
    officer[name]=0
    officer.loc[(officer['appointed_year'] <= year) & # if the police join on and before that year
       (officer['resignation_year'] >= year), # if the police left on and after that year
       name] = 1
    return(officer[name])
 
for i in range(2007, 2018):
    officer[f'served_{i}']=new_served_year_column(i, f'served_{i}')
```

```python
# Now loading the new dataset

import pandas as pd
officerPOP = pd.read_csv("officer_v2_Population_2007-2017.csv")
officerPOP.head(5)
```

```python
# create a function group the gener, race in each year population
def gender_race_POPcount(year):
    YearGroup = officerPOP.loc[officerPOP[f'served_{year}'] == 1]
    YearTable = YearGroup.groupby(['gender', 'race']).size().reset_index(name='population_counts')
    table = YearTable.assign(Year=f'{year}')
    return(table)
#gender_race_POPcount(2008)   --- try to use year 2008
```

```python
# Use the function created above to loop over 2007 - 2017 and then create a population dataframe
table_list=[]
for i in range(2007, 2018):
    individual_table = gender_race_POPcount(i) 
    table_list.append(individual_table)
population = pd.concat(table_list)
population
```
